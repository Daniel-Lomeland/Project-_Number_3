```python
import pandas as pd
from sqlalchemy import create_engine
import pymysql
pymysql.install_as_MySQLdb()
```

# Store Shelter CSV into DataFrame

```python
csv_file = "Resources/aac_shelter_outcomes.csv"
shelter_data_df = pd.read_csv(csv_file)
shelter_data_df.head()
```

# Create new data for Animal Type with select columns

```python
new_animal_type_data_df = shelter_data_df[['animal_type']].copy()
new_animal_type_data_df.head()
new_animal_type_data_df.drop_duplicates(['animal_type'],inplace=True,keep='first')

new_animal_type_data_df['Animal_type_Id'] = range(1,1+len(new_animal_type_data_df))
new_animal_type_data_df.head()
cols = new_animal_type_data_df.columns.tolist()
cols = cols[-1:] + cols[:-1]
new_animal_type_data_df = new_animal_type_data_df[cols]
new_animal_type_data_df.head(10)
```

# Create new data for Breed with select columns

```python
new_breed_data_df = shelter_data_df[['breed']].copy()
new_breed_data_df.head()
new_breed_data_df.drop_duplicates(['breed'],inplace=True,keep='first')

new_breed_data_df['Breed_Id'] = range(1,1+len(new_breed_data_df))
new_breed_data_df.head()
cols = new_breed_data_df.columns.tolist()
cols = cols[-1:] + cols[:-1]
new_breed_data_df = new_breed_data_df[cols]
#new_breed_data_df.rename(columns={'breed':'breed_name'},inplace=True)
new_breed_data_df.head(10)
```

# Create new data for Color with select columns

```python
new_color_data_df = shelter_data_df[['color']].copy()
new_color_data_df.head()
new_color_data_df.drop_duplicates(['color'],inplace=True,keep='first')

new_color_data_df['Color_Id'] = range(1,1+len(new_color_data_df))
new_color_data_df.head()
cols = new_color_data_df.columns.tolist()
cols = cols[-1:] + cols[:-1]
new_color_data_df = new_color_data_df[cols]
#new_color_data_df.rename(columns={'color':'color_name'},inplace=True)
new_color_data_df.head(10)
```

# Create new data for Outcome Type with select columns

```python
new_outcome_type_data_df = shelter_data_df[['outcome_type']].copy()
new_outcome_type_data_df.head()
new_outcome_type_data_df.drop_duplicates(['outcome_type'],inplace=True,keep='first')
new_outcome_type_data_df = new_outcome_type_data_df.dropna()

new_outcome_type_data_df['Outcome_type_Id'] = range(1,1+len(new_outcome_type_data_df))
new_outcome_type_data_df.head()
cols = new_outcome_type_data_df.columns.tolist()
cols = cols[-1:] + cols[:-1]
new_outcome_type_data_df = new_outcome_type_data_df[cols]
#new_outcome_type_data_df.rename(columns={'outcome_type':'outcome_type_name'},inplace=True)
new_outcome_type_data_df.head(10)
```

# Create new data for Sex Upon Outcome  with select columns

```python
new_sex_data_df = shelter_data_df[['sex_upon_outcome']].copy()
new_sex_data_df.head()
new_sex_data_df.drop_duplicates(['sex_upon_outcome'],inplace=True,keep='first')

new_sex_data_df['Sex_Id'] = range(1,1+len(new_sex_data_df))
new_sex_data_df.head()
cols = new_sex_data_df.columns.tolist()
cols = cols[-1:] + cols[:-1]
new_sex_data_df = new_sex_data_df[cols]
new_sex_data_df = new_sex_data_df.dropna()
#new_sex_data_df.rename(columns={'sex_upon_outcome':'sex_name'},inplace=True)
new_sex_data_df.head(10)
```

# Create ACC Shelter Dataframe

```python
result_acc_shelter = pd.merge(shelter_data_df,new_animal_type_data_df,on=['animal_type'])
result_acc_shelter_breed = pd.merge(result_acc_shelter,new_breed_data_df,on=['breed'])
result_acc_shelter_color = pd.merge(result_acc_shelter_breed,new_color_data_df,on=['color'])
result_acc_shelter_outcome_type = pd.merge(result_acc_shelter_color,new_outcome_type_data_df,on=['outcome_type'])
result_acc_shelter_sex_upon_outcome = pd.merge(result_acc_shelter_outcome_type,new_sex_data_df,on=['sex_upon_outcome'])
result_acc_shelter_sex_upon_outcome.head(5)
```

# Drop Columns

```python
result_shelter_df = result_acc_shelter_sex_upon_outcome.drop(columns=['monthyear','outcome_subtype','animal_type',\
                        'breed','color','sex_upon_outcome','outcome_type'])
column = list(result_shelter_df.head(0))
print(column)
result_shelter_df.head()
```

# Move ANIMAL_ID to front

```python
animal_id = result_shelter_df['animal_id']
result_shelter_df.drop(labels=['animal_id'], axis=1,inplace = True)
result_shelter_df.insert(0, 'animal_id', animal_id)
result_shelter_df.head(10)
```

# Rename Columns

```python
result_shelter_df.rename(columns={'date_of_birth':'dob',
                               'datetime':'date_shelter',
                               'name':'name_of_animal'},inplace=True)
result_shelter_df.head()
```

# Connect to local database

```python
rds_connection_string = "root:tito2410@127.0.0.1/shelter_db"
engine = create_engine(f'mysql://{rds_connection_string}')
```

# Check for tables

# Load animal_type Table to MySql

```python
new_animal_type_data_df.rename(columns={'animal_type':'animal_type_name'},inplace=True)
new_animal_type_data_df.to_sql(name='animal_type', con=engine, if_exists='append', index=False)
pd.read_sql_query('select * from animal_type', con=engine).head()
```

# Load Breed Table to MySql

```python
new_breed_data_df.rename(columns={'breed':'breed_name'},inplace=True)
new_breed_data_df.to_sql(name='breed', con=engine, if_exists='append', index=False)
pd.read_sql_query('select * from breed', con=engine).head()
```

# Load Color Table to MySql

```python
new_color_data_df.rename(columns={'color':'color_name'},inplace=True)
new_color_data_df.to_sql(name='color', con=engine, if_exists='append', index=False)
pd.read_sql_query('select * from color', con=engine).head()
```

# Load Outcome Type Table to MySql

```python
new_outcome_type_data_df.rename(columns={'outcome_type':'outcome_type_name'},inplace=True)
new_outcome_type_data_df.to_sql(name='outcome_type', con=engine, if_exists='append', index=False)
pd.read_sql_query('select * from outcome_type', con=engine).head()
```

# Load Sex Upon Outcome Table to MySql

```python
new_sex_data_df.rename(columns={'sex_upon_outcome':'sex_name'},inplace=True)
new_sex_data_df.to_sql(name='sex_upon_outcome', con=engine, if_exists='append', index=False)
pd.read_sql_query('select * from sex_upon_outcome', con=engine).head()
```

# Load AAC Shelter Outcome Table to MySql

```python
result_shelter_df['DOB'] = pd.to_datetime(result_shelter_df['DOB'])
result_shelter_df['date_shelter'] = pd.to_datetime(result_shelter_df['date_shelter'])
result_shelter_df.drop_duplicates(['animal_id'],inplace=True,keep='first')
result_shelter_df.to_sql(name='acc_shelter_outcome', con=engine, if_exists='append', index=False)
pd.read_sql_query('select * from acc_shelter_outcome', con=engine).head()
```
